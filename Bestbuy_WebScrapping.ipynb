```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Launch the browser
browser = Browser('chrome')
```

```python
url = 'https://www.bestbuy.ca/en-ca/category/televisions/21344'
# Send the browser to the URL
browser.visit(url)

# Save the HTML from the browser
html = browser.html

# Create a BeautifulSoup object from the HTML
html_soup = soup(html, 'html.parser')
```

```python
import time
# Scroll down to load more content
scroll_pause_time = 2
scroll_count = 15  # Number of times to scroll down, adjust as needed

for _ in range(scroll_count):
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause_time)

# Save the HTML from the browser
html = browser.html

# Create a BeautifulSoup object from the HTML
html_soup = soup(html, 'html.parser')


# Tv_description = html_soup.find_all('div', class_= 'productItemName_3IZ3c')
# Tv_description


# Find all TV names
Tv_name_elements = html_soup.find_all('div', class_='productItemName_3IZ3c')

# Extract TV names and store them in a list
Tv_names = [Tv_name.text.strip() for Tv_name in Tv_name_elements]

# Display the TV names
for name in Tv_names:
    print(name)

```

```python
# Tv_prices= html_soup.find_all('div', class_='productPricingContainer_3gTS3')
# Tv_prices
 

# Find all TV prices
Tv_price_elements = html_soup.find_all('div', class_='productPricingContainer_3gTS3')

# Extract TV prices and store them in a list
Tv_prices = [price.text.strip() for price in Tv_price_elements]

# Display the TV prices
for price in Tv_prices:
    print(price)

```
