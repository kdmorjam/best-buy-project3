```python
# import Dependencies
import pandas as pd
import numpy as np
import re
```

```python
# creating the path
csv_path = "C:/Users/olaka/OneDrive/Documents/GitHub/project 3/bestbuy_data.csv"
Bestbuy_Webscraping= pd.read_csv(csv_path)
Bestbuy_Webscraping.head()
```

```python
# Create a new DataFrame
new_dfWebscraping = Bestbuy_Webscraping.copy()


#cleaning the reviwe column

new_dfWebscraping['reviews'] = new_dfWebscraping['reviews'].replace('Bring the cinematic experience home with Sony.', np.nan)

#cleaning the price column
new_dfWebscraping['Price'] = new_dfWebscraping['Price'].replace('Gift with purchase', np.nan)
new_dfWebscraping['Price'] = new_dfWebscraping['Price'].replace('Learn more', np.nan)


# cleaning the Discount column
new_dfWebscraping['Discount'] = new_dfWebscraping['Discount'].str.replace('SAVE ', '')
new_dfWebscraping['Discount'] = new_dfWebscraping['Discount'].replace('Available to ship', np.nan)
new_dfWebscraping['Discount'] = new_dfWebscraping['Discount'].replace('Sold out online', np.nan)
new_dfWebscraping['Discount'] = new_dfWebscraping['Discount'].replace('Available online only', np.nan)
new_dfWebscraping['Discount'] = new_dfWebscraping['Discount'].replace('Available for backorder', np.nan)
new_dfWebscraping['Discount'] = new_dfWebscraping['Discount'].replace('Not sold online', np.nan)


# Remove 'Plus $XX.XX EHF' from the Discount column
new_dfWebscraping['Discount'] = new_dfWebscraping['Discount'].str.replace(r'Plus \$(\d+\.\d+) EHF', r'\1')

```

```python
# Rename columns to make them more understandable
new_dfWebscraping = new_dfWebscraping.rename(columns={"":"Index",
                                               "description": "Description",
                                               "reviews": "Reviews",
                                               "Price": "Price",
                                               "Discount": "Discount",
                                               "Brand": "Brand",
                                                "TV Type" :"TV_Type",
                                                  "TV Size": "TV_Size",
                                               "TV Pixels" :"TV_Pixels",                                           
                                   })
```

```python
# Extracting Brand
# new_dfWebscraping['Brand'] = new_dfWebscraping['description'].str.extract(r'([A-Za-z]+)\s?\d+"')
# new_dfWebscraping['Brand'] = new_dfWebscraping['description'].str.extract(r'(Amazon Echo)')
tv_brand= ['Amazon Echo', 'Samsung', 'LG', 'Naxa', 'Sony','Hisense', 'Amazon fire TV','Toshiba','TCL 75', 'Insignia','SAMSUNG' ,'VIZIO', 'ONN.70', 'TCL 6Series 65', 'Philips 75', 'RCA 32','ONN 65â€', 'JVC 58']
# pattern = '|'.join(brand_names)
# new_dfWebscraping['Brand'] = new_dfWebscraping['description'].str.extract(f'({pattern})')
# Function to extract TV brand from description
def extract_tv_brand(description):
    for brand in tv_brand:
        if re.search(brand, description, re.IGNORECASE):
            return brand
    return None
new_dfWebscraping['Brand'] = new_dfWebscraping['Description'].apply(extract_tv_brand)

# Get TV Type
new_dfWebscraping['TV_Type'] = new_dfWebscraping['Description'].str.extract(r'(LCD|LED|OLED|QNED)')

# Extract TV size using regular expression
new_dfWebscraping['TV_Size'] = new_dfWebscraping['Description'].str.extract(r'(\d+)"')

#TV Pixels
new_dfWebscraping['TV_Pixels'] = new_dfWebscraping['Description'].str.extract(r'(Ultra HD|4K|1080p|720p)')

# Print the updated DataFrame
new_dfWebscraping
```
