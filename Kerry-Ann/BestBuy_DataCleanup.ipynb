```python
#Import dependencies
import pandas as pd
import numpy as np
import re
```

```python
# File to Load 
bestbuy_data_to_load = "data_csv"

# Read Data File and store into Pandas DataFrames
bestbuy_data = pd.read_csv(bestbuy_data_to_load,)

# Combine the data into a single dataset.  
bestbuy_data.head()
```

```python
bestbuy_data = bestbuy_data.drop("Unnamed: 0", axis=1)
```

```python
bestbuy_data.dtypes
```

```python
#bestbuy_data['New Discount'] = bestbuy_data['Discount'].str.extract(r'\$(\d+(?:\.\d+)?)')
bestbuy_data['New Price'] = bestbuy_data['Price'].str.extract(r'\$(\d+(?:\.\d+)?)')
```

```python
bestbuy_data.head()
```

```python
bestbuy_data = bestbuy_data.drop("New Discount", axis=1)
```

```python
bestbuy_data["Price"] =bestbuy_data["New Price"]
```

```python
bestbuy_data.head()
```

```python
bestbuy_data['New Discount'] = bestbuy_data['Discount'].str.replace('SAVE', '')
```

```python
bestbuy_data.head()
```

```python
bestbuy_data["Discount"] =bestbuy_data["New Discount"]
```

```python
bestbuy_data.head()
```

```python
# Extract brand name using regular expression
bestbuy_data['Brand'] = bestbuy_data['Description'].str.extract(r'([A-Za-z]+)\s?\d+"')
```

```python
bestbuy_data
```

```python
bestbuy_data["Discount"] = bestbuy_data["Discount"].fillna(0)
```

```python
bestbuy_data
```

```python
# Reorder the columns
new_column_order = ['Brand', 'Description', 'Price', 'Discount']
bestbuy_data = bestbuy_data.reindex(columns=new_column_order)
```

```python
bestbuy_data
```

```python

# Extract TV types using regular expression
bestbuy_data['TV Type'] = bestbuy_data['Description'].str.extract(r'(LCD|LED|OLED|QNED)')
bestbuy_data
```

```python
bestbuy_data['TV Pixels'] = bestbuy_data['Description'].str.extract(r'(Ultra HD|4K|1080p|720p)')
```

```python
bestbuy_data.head(50)
```

```python
# Extract TV size using regular expression
bestbuy_data['TV Size'] = bestbuy_data['Description'].str.extract(r'(\d+)"')
```

```python
bestbuy_data.head(50)
```

```python
bestbuy_data['Price'] = '$' + bestbuy_data['Price'].apply(lambda x: '$' + x)
```

```python
bestbuy_data.head(50)
```

```python
bestbuy_data.dtypes
```

```python
#Convrt column types
bestbuy_data['Price'] = bestbuy_data['Price'].str.replace('$', '').astype(float)
bestbuy_data['Discount'] = bestbuy_data['Discount'].str.replace('$', '').str.replace(',', '').astype(float)
#Remove NaN before converting TV Size column
bestbuy_data['TV Size'] = bestbuy_data['TV Size'].fillna(0)
bestbuy_data['TV Size'] = bestbuy_data['TV Size'].astype(int)
```

```python
bestbuy_data.dtypes
```

```python
bestbuy_data.head(50)
```

```python
bestbuy_data.to_csv("tv_listing", index=False)
```
